A function to iterate through all the folders and return course name and an array of term year, instructor name, syllabus path

```python
import os
def calcpath(path):
    dir_path = os.path.dirname(os.path.realpath(path)) 
    ##print(dir_path)
    filepdf = []
    rootpdf = []
    filedoc = []
    rootdoc = []
    filedocx = []
    rootdocx = []
    path = []
    
    for root, dirs, files in os.walk(dir_path): 
        for file in files:  
  
        # change the extension from '.mp3' to  
        # the one of your choice. 
            if file.endswith('.pdf'):
            #print (root+'/'+str(file))
                a = root+'/'+str(file)
                filepdf.append(file)
                rootpdf.append(a)
            ##print(a)
        
            elif file.endswith('.doc'):
            #print (root+'/'+str(file))
                b = root+'/'+str(file)
                filedoc.append(file)
                rootdoc.append(b)
        
            elif file.endswith('.docx'):
            #print (root+'/'+str(file))
                c = root+'/'+str(file)
                filedocx.append(file)
                rootdocx.append(c)
    
    roots = np.concatenate((rootpdf, rootdoc, rootdocx))
    files = np.concatenate((filepdf, filedoc, filedocx))
    
    arr =[]
    arrnew = []
    dftest = pd.DataFrame({'Subject_Catalog':[], 'Instructor_Name':[], 'Term_Offered':[], 'Syllabus_Link':[], 'Array_path': []})
    finalarr = []
    
    for root in roots:
        arr = root.split('/')
        path = root
        name = arr[1]
        arrnew = name.split('_')
        #course_name = arrnew[0]
        newlst = [l.replace('.pdf','') for l in arrnew ]  
        newlst1 = [l.replace('.docx','') for l in newlst ] 
        newlst2 = [l.replace('.doc','') for l in newlst1 ] 
        newlst2.append(root)
        if len(newlst2) == 4:
            Subject_Catalog = newlst2[0]
            Instructor_Name = newlst2[1]
            Term_Offered = newlst2[2]
            Syllabus_Link = root
        else:
            Subject_Catalog = arrnew[0]
            Instructor_Name = "Not available"
            Term_Offered = "Not available"
            Syllabus_Link = "Not available"
            newlst2 = []
        row = {'Subject_Catalog': Subject_Catalog, 'Instructor_Name':Instructor_Name, 'Term_Offered':Term_Offered, 'Syllabus_Link':Syllabus_Link, 'Array_path': newlst2 }
        dftest = dftest.append(row, ignore_index = True)
              
    return dftest
```
