**A function to iterate through three arrays (rootpdf,rootdoc,rootdocx) and check if the syllabus file is available.
This function accepts two parameters param = course number, dep = course name. 
An empty array is defined in the beginning of the function. 
If course number and course name (lower case/upper case) or Combination of course name(upper case)+ course number exists in any of the three arrays of paths, that path is appended to the empty array. 
At the end, the array contains all the roots of the syllabus**

```python
def calcpath(param,dep):
    arr =[]
    arrnew = []
    val = param
    dep = dep
    deplower = str.lower(dep)
    comb = dep+val
    
    for root in rootpdf:
        if val in root:
            if dep in root:
                arr.append(root)
            if deplower in root:
                arr.append(root)
        elif comb in root:
            arr.append(root)
    for root in rootdoc:
        if val in root:
            if dep in root:
                arr.append(root)
            if deplower in root:
                arr.append(root)
        elif comb in root:
            arr.append(root)
    for root in rootdocx:
        if val in root:
            if dep in root:
                arr.append(root)
            if deplower in root:
                arr.append(root)
        elif comb in root:
            arr.append(root)
    
    
    return arr
```

**Function to return the file name for each path**

```python
def calcfile(path):
    arr= []
    path = path
    for file in path:
        for root in filepdf:
            if root in file:
                arr.append(root)
        for root in filedoc:
            if root in file:
                arr.append(root)
        for root in filedocx:
            if root in file:
                arr.append(root)
    return arr
    
            
```

**Function to split the file name with "_" as a separator to extract course name, instructor name, quarter and year**

```python
def separator(file):
    file = file
    lst = ()
    newlst =()
    newlst1 = ()
    newlst2 = ()
    arr = []
    
    for filename in file:
        lst = filename.split("_")
        newlst = [l.replace('.pdf','') for l in lst ]  
        newlst1 = [l.replace('.docx','') for l in newlst ] 
        newlst2 = [l.replace('.doc','') for l in newlst1 ] 
        arr.append(newlst2)
         
    return arr
```

**A function to check if the syllabus for a given course name and number is available in the folder. If yes, True is returned else False**

```python
def chkstr(param,dep):
    res = False
    param = param
    dep = dep
    deplower = str.lower(dep)
    comb = dep+param
  
    ##Check if the param, dep, and combination of dep+param exist in pdf roots.
    res = param in temp
    if res == True:
        res = (dep in temp or deplower in temp)
    elif res == False:
        res = comb in temp
    ##Check if the param, dep, and combination of dep+param exist in doc roots.   
    if res == False:
        res = param in tempdoc
        if res == True:
            res = (dep in tempdoc or deplower in tempdoc)
        elif res == False:
            res = comb in tempdoc
    ##Check if the param, dep, and combination of dep+param exist in docx roots.   
    if res == False:
        res = param in tempdocx
        if res == True:
            res = (dep in tempdocx or deplower in tempdocx)
        elif res == False:
            res = comb in tempdocx
    return res
```
